Sentence boundary disambiguation (SBD), also known as sentence breaking, sentence boundary detection, and sentence segmentation, is the problem in natural language processing of deciding where sentences begin and end. Natural language processing tools often require their input to be divided into sentences; however, sentence boundary identification can be challenging due to the **potential ambiguity of punctuation marks**. In written English, a period may indicate the end of a sentence, or may denote an **abbreviation, a decimal point, an ellipsis, or an email address**, among other possibilities.

* https://grammar.yourdictionary.com/punctuation/what/fourteen-punctuation-marks.html
* https://aclanthology.org/J97-2002.pdf
* https://dbpedia.org/page/Sentence_boundary_disambiguation


### Steps followed:
* 1. First split sentances into words based on space.
* 2. Then based on the terminating punctuations (.?!) split the sentances into **leftword + punctuation + rightword** and add ending marker at the potential boundary places
* 3. Then past the split sentances through rules and check if the boundaries are correct. Store the correct boundaries in a 
     boundary list
* 4. Depending on the final boundary list obtained add boundary marker
* 5. Split the string based on the boundary marker


## Helper Functions

Making helper functions for checking certain conditions like:
* terminating punctuations
* numeric character
* double quotes
* abbreviation type 1
* abbreviation type 2
* word is capital 
* word starting and ending with double quotes
* checking url
* checking email
* check elipsis
* check decimals

```python
# terminating punctuations

def isTermPunct(character):
    punct = [".","!","?"]
    flag=0
    for i in punct:
#         print(punct[i]) 
       if(i==character):
            flag=1

    if(flag==1):
        return True
    else:
        return False
    
    
# numbers

def checkNumChar(character):
    numchar = ['#','$','%','+','>','<','/']
    flag=0
    for i in numchar:
        if (character==i):
            flag=1
            break
        else:
            flag=0
    if(flag==1):
        return True
    else:
        return False
    
    
# double quotes

def isDoubleQuotes(character):
    if(character=='"'):
        return True
    else:
        return False
        




#Abbreviation

# Type 1 abbreviation. For eg. Dr., Mr. Mrs. etc.
def is_abbr1(word):
    flag=0
    abbr1 = ["Dr.","Mr.","Mrs.","Phd.","Prof.","Drs.","etc.","i.e.","eg.","Rs."]
    for i in range(len(abbr1)):
#         print(abbr1[i]) 
        if(abbr1[i]==word):
#             print(word)
            flag=1
    if(flag==1):
        return True
    else:
        return False
        
# Type 2 abbreviation. For eg. S.A.R. , R.A.D. etc.        
def is_abbr2(word):
    for i in range(len(word)):
        if(word[i].isupper()) and (word[i+1]=="."):
            return True
        else:
            return False

        
# is word capital

def isWordCap(word):
    if(word[0]).isupper():
        if(len(word)==1):
            return True
        for i in range(1,len(word)):
            if(word[i].islower()):
                return True
            else:
                return False

            
# Words ending with double quotes
def endQuotes(word):
    n = len(word)
    if(isDoubleQuotes(word[n-1]) and isTermPunct(word[n-2])):
        return True
    else:
        return False

# Words starting with double quotes
def startQuotes(word):
    n = len(word)
    if(isDoubleQuotes(word[0])):
        return True
    else:
        return False
    
    
# to check urls
def checkUrl(str):
 
    # Regex to check valid URL
    regex = ("((http|https)://)(www.)?" +
             "[a-zA-Z0-9@:%._\\+~#?&//=]" +
             "{2,256}\\.[a-z]" +
             "{2,6}\\b([-a-zA-Z0-9@:%" +
             "._\\+~#?&//=]*)")
     
    # Compile the ReGex
    p = re.compile(regex)

    if (str == None):
        return False

    if(re.search(p, str)):
        return True
    else:
        return False

# to check Emails
def checkEmail(email):
    regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
 
    if(re.fullmatch(regex, email)):
        return True
 
    else:
        return False

# check elipsis ...
def checkElipsis(word):
    
    n = len(word)
    if(word[n-1]=="." and word[n-2]=="." and word[n-3]=="."):
        return True
    if(isDoubleQuotes(word[n-1]) and word[n-2]=="." and word[n-3]=="." and word[n-4]=="."):
        return True
    else:
        return False
    
    
# check decimal numbers    
def checkDecimal(word):
    r = re.compile(r"^\d*[.,]?\d*$")
    if(r.match(word)):
        return True
    else:
        return False
```

## Rules function

making a rules function to check all the rules applied

```python
def checkRules(strlist):
    flag=0
#     print(strlist[2])
#     print(isWordCap(strlist[2]))
    if (isTermPunct(strlist[1]) and (is_abbr1(strlist[0]))==False): # left not abbr + terminating punct + Word cap
        if(isWordCap(strlist[2]) or strlist[2]=='$$'): #if right word is cap or $$(end word)
            flag=1
    if(endQuotes(strlist[0])) and (isWordCap(strlist[2])):   #leftword + term punct + end quotes + capital word 
        flag=1
    if(isTermPunct(strlist[1])) and (startQuotes(strlist[2])):   #leftword + termpunct + startquotes
        flag=1
    if(checkElipsis(strlist[0]) and isWordCap(strlist[2])):     #leftword with elipsis + punct + rightword cap/small -> no
        flag=1
    if(checkUrl(strlist[0]) or checkEmail(strlist[0])):     #checking url and email in string
        flag=0
    if(checkDecimal(strlist[0])):    #checking decimal
        flag=0    
#     print(flag)
    if(flag):
        return True
    else:
        return False
```

## Sentance helper functions

making sentance helper functions like:
* splitting the paragraph to list for making initial list based on **Terminating punctuations** 
* converting list to strings
* initial splitter of string which combines boith the above functions

```python
## split paragraph to list

def splitParaList(string):
    final=[]
    lis = []
    s = string.split(" ")
    n=len(s)
    for i in range(len(s)):
#         print(s[i])
        for j in s[i]:
            if(isTermPunct(j)):
                lis.append(s[i])
                lis.append(j)
                s[i] = s[i]+"$$"
                if(i==(len(s)-1)):
                    lis.append("$$")
                else:
                    lis.append(s[i+1])

    return lis, s


## convert list to string
def listToString(s): 
    
    # initialize an empty string
    str1 = "" 
    
    # traverse in the string  
    for i in range(len(s)): 
        if(i!=len(s)-1):
            str1 += s[i]+" " 
        else:
            str1+=s[i]
    
    # return string  
    return str1 

## Splitting the initial list based on terminating punctuations

def initialSplitter(string):
    finalis, finalstr = splitParaList(string)    
    n=3
    final = [finalis[i * n:(i + 1) * n] for i in range((len(finalis) + n - 1) // n )] 
    return final
```

## Checking sentance boundary

* first we find the boundary by giving the initial strings list as input
* we get output of the boundary words 
* then we split the final sentance based on the boundary and add sentance markers '$$'

```python
## finding sentance boundary

def findBoundary(strlist):
    list2 = []
    for i in final:
        if(checkRules(i)):
            list2.append(i)
    boundary = []
    for i in list2:
        boundary.append(i[0])
        
    return boundary

## converting to sentance
def finalSentance(string, boundary):
    s = string.split(" ")
    n=len(s)
    m=len(boundary)
    for i in range(len(s)):
        k=0
        while(k<m):
            if(s[i]==boundary[k]):
                s[i]+="$$"
                break
            k=k+1
    return listToString(s)
```
